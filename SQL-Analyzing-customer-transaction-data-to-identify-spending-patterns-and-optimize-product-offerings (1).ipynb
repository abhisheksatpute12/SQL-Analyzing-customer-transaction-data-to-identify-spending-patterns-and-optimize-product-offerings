```python
query2="""
SELECT 
    Customer_ID,
    Customer_Name,
    COUNT(Product_ID) AS Sales_Volume
FROM 
    retail_superstore 
GROUP BY 
    Customer_ID,
    Customer_Name
ORDER BY 
    Sales_Volume DESC
LIMIT 10
"""
result = pd.read_sql_query(query2, conn)
result
```

2.What are the different customer segments, and how do they differ in purchasing behavior?

```python
query3="""
SELECT 
    Segment, 
    COUNT(DISTINCT Customer_ID) AS Number_of_Customers,
    COUNT(Order_ID) AS Total_Orders,
    SUM(Sales) AS Total_Sales,
    AVG(Sales) AS Average_Order_Value,
    SUM(Sales) / COUNT(DISTINCT Customer_ID) AS Average_Sales_Per_Customer
FROM 
    retail_superstore
GROUP BY 
    Segment
ORDER BY 
    Total_Sales DESC;
"""
result = pd.read_sql_query(query3, conn)
result
```

# Insights
- Number Of Customers in Consumer Segment( 409 Distinct Customer) Leading to their highest total sales( 114061 USD)
- Consumer segment generates the highest total sales and revenue.


3.What are the average order values and order frequencies by customer?

```python
query3="""
SELECT 
    Customer_ID,
    Customer_Name,
    COUNT(Order_ID) AS Order_Frequency,
    SUM(Sales) AS Total_Sales,
    AVG(Sales) AS Average_Order_Value
FROM 
    retail_superstore
GROUP BY 
    Customer_ID, Customer_Name
ORDER BY 
    Total_Sales DESC;

"""
result = pd.read_sql_query(query3, conn)
result
```

# Insights
- Order Frequency: How frequently each customer places orders.
- Average Order Value: The typical value of each order for a customer, which gives insight into their purchasing habits and potential value to the business.

Are there regional differences in product preferences?

```python
query14="""
SELECT 
    Region, 
    Product_ID, 
    Product_Name, 
    Total_Sales
FROM (
    SELECT 
        Region, 
        Product_ID, 
        Product_Name, 
        SUM(Sales) AS Total_Sales,
        ROW_NUMBER() OVER (PARTITION BY Region ORDER BY SUM(Sales) DESC) AS Sales_Rank
    FROM 
        retail_superstore
    GROUP BY 
        Region, Product_ID, Product_Name
) AS RankedProducts
WHERE 
    Sales_Rank <= 5  -- Selecting top 5 products in each region
ORDER BY 
    Region, Sales_Rank;

"""
result = pd.read_sql_query(query14, conn)
result
```

# Insights
- Discovered regional product preferences to tailor marketing and product offerings according to local demand.
- Top 5 products which has highest sales is given.

# Seasonal and Time-Based Analysis <a id="i1"></a>

What are the peak sales months or seasons?

```python
query="""
SELECT 
    CASE 
        WHEN CAST(substr(Order_Date, 4, 2) AS INTEGER) IN (12, 1, 2) THEN 'Winter'
        WHEN CAST(substr(Order_Date, 4, 2) AS INTEGER) IN (3, 4, 5) THEN 'Spring'
        WHEN CAST(substr(Order_Date, 4, 2) AS INTEGER) IN (6, 7, 8) THEN 'Summer'
        WHEN CAST(substr(Order_Date, 4, 2) AS INTEGER) IN (9, 10, 11) THEN 'Fall'
    END AS Season,
    SUM(Sales) AS Total_Sales
FROM 
    retail_superstore
GROUP BY 
    Season
ORDER BY 
    Total_Sales DESC;

"""
result = pd.read_sql_query(query, conn)
result
```

```python
query="""
SELECT 
    CAST(substr(Order_Date, 4, 2) AS INTEGER) AS Month,
    SUM(Sales) AS Total_Sales
FROM 
    retail_superstore
GROUP BY 
    Month
ORDER BY 
    Total_Sales DESC;
"""
result = pd.read_sql_query(query, conn)
result
```

# Insights
- At Fall, Total Sales is highest
- At November , Total Sales is highest

```python
query="""
SELECT 
    Ship_Mode,
    AVG(julianday(
        substr(Ship_Date, 7, 4) || '-' || substr(Ship_Date, 4, 2) || '-' || substr(Ship_Date, 1, 2)
    ) - julianday(
        substr(Order_Date, 7, 4) || '-' || substr(Order_Date, 4, 2) || '-' || substr(Order_Date, 1, 2)
    )) AS Avg_Delivery_Time
FROM 
    retail_superstore
WHERE 
    Ship_Date IS NOT NULL AND Order_Date IS NOT NULL
GROUP BY 
    Ship_Mode
ORDER BY 
    Avg_Delivery_Time ASC;

"""
result = pd.read_sql_query(query, conn)
result
```
